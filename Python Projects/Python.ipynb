\n = split over lines
\t = tabs/indents the line
\' or \" = if you want to use single quotes inside '' or double quotes inside ""
\ = same line, even if split over multiple lines.

```python
splitString = "This string has been\nsplit over\nseveral\nlines"
print(splitString)

tabbedString = "1\t2\t3\t4\t5"
print(tabbedString)

print('The pet shop owner said "No, no, \'e\'s uh,...he\'s resting".')
print("The pet shop owner said \"No, no, 'e's uh,...he's resting\".")

print("""The pet shop owner said "No, no, 'e's uh,...he's resting". """)

anotherSplitString = """This string has been 
split over 
several 
lines."""

lastSplitString = """This string has been \
split over \
several \
lines."""

print(anotherSplitString)
print(lastSplitString)
```

```python
# Mini Challenge
# Print "we win" from the variable parrot. 

# Easy way
print(parrot[3] + parrot[4] + " " + parrot[3] + parrot[6] + parrot[8])

# A little harder way

stringOne = []
for char in (parrot):
    if char in "wein":
        stringOne.append(char)
    else:
        continue

# ['w', 'e', 'i', 'n', 'e']

del stringOne[4]
print(stringOne)
# ['w', 'e', 'i', 'n']

# rearranging the list
stringOne.append(" ")
stringOne.append("w")

# ['w', 'e', 'i', 'n', ' ', 'w']

stringOne[2], stringOne[4] = stringOne[4], stringOne[2]

# ['w', 'e', ' ', 'n', 'i', 'w']

stringOne[3], stringOne[5] = stringOne[5], stringOne[3]


print("".join(stringOne))

```

```python
a = 15
b = 8

print(a + b) # Add
print(a - b) # Subtract
print(a * b) # Multiply
print(a / b) # Divide (float)
print(a // b) # Integer division, rounded down towards minus infinity
print(a % b) # 0 modulo: the remainder after integer division 
```
