Indexing    Slicing
   []   [start:End:Step]

```python
                    #1
#parrot= "012345678901234"
parrot = "Norwegian Blue"
print(parrot)

print(parrot[3]) # prints the third letter, starting from 0. 
# Python starts counting from 0
print(parrot[-3]) # Negative indexing starts from the last letter 

# Substrings
print(parrot[10:]) # Blue
print(parrot[:9]) #Norwegian
print(parrot[::2]) # Every second character -> NreinBu
print(parrot[-4:14]) # Negative slicing
print(parrot[8::-1]) # negative step

raceCar = "Race Car"
print(raceCar[::1]) # Should print Race Car
print(raceCar[::-1]) # Should print raC ecar (This is also an idiom: reversing the sequence)
```

```python
# Mini Challenge
# Print "we win" from the variable parrot. 

# Easy way using indexing
print(parrot[3] + parrot[4] + " " + parrot[3] + parrot[6] + parrot[8])

# A little harder 

stringOne = []
for char in (parrot):
    if char in "wein":
        stringOne.append(char)
    else:
        continue

# ['w', 'e', 'i', 'n', 'e']

del stringOne[4]
print(stringOne)
# ['w', 'e', 'i', 'n']

# rearranging the list
stringOne.append(" ")
stringOne.append("w")

# ['w', 'e', 'i', 'n', ' ', 'w']

stringOne[2], stringOne[4] = stringOne[4], stringOne[2]

# ['w', 'e', ' ', 'n', 'i', 'w']

stringOne[3], stringOne[5] = stringOne[5], stringOne[3]


print("".join(stringOne))

```

```python
# Most optimal way using states. 
parrot = "Norwegian Blue"
out = []
state = 0

for char in parrot:
    if state == 0:
        if char == "w":
            out.append("w")
            state = 1
    
    elif state == 1:
        if char == "e":
            out.append("e")
            out.append(" ")
            out.append("w")
            state = 2
        
    elif state == 2:
        if char == "i":
            out.append("i")
            state = 3

    elif state == 3:
        if char == "n":
            out.append("n")
            state = 4
            break

print("\n".join(out))
```

## Guessing Game

```python
import random 

answer = random.randint(1, 10)
count = 0

print("Please guess a number between 1 and 10: ")

while True:
    guess = int(input())
    if guess > answer:
        print("Please guess lower!")
        count += 1
        continue
    elif guess < answer:
        print("Please guess higher!")
        count += 1
        continue
    else:
        print(f"Well done, you guessed it in {count} tries!")
        break

```
