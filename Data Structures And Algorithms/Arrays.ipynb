# Static Arrays, Dynamic Arrays & Strings.

Arrays

```python
A = [1, 2, 3]

print(A)
```

```python
# Append - Insert a element at the end of an array - on average O(1)
A.append(5)

print(A)
```

```python
# Pop - Delete a element at the end of an array - O(1)
A.pop()

print(A)
```

```python
# Insert (not at the end) - O(n)
A.insert(2, 5)

print(A)
```

```python
# Modify an element - O(1)
A[0] = 7 

print(A)
```

```python
# Accessing a element given index i - O(1)
print(A[2])
```

```python
# Checking if an array has an element - O(n)
if 7 in A:
    print(True)
```

```python
# Check length - O(1)
print(len(A))
```

Strings

```python
# Append to the end of string - O(n)
s = "Hello"

b = s + " World"

print(b)
```

```python
# Check if something is in string - O(n)
if 'f' in s:
    print(True)
```

```python
# Access positions - O(1)
print(s[4])
```

```python
# Check length - O(1)
print(len(s))
```
